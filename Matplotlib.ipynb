```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
plt.plot(x, y, 'r')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('To Plot The Signal Of y=x**2')
plt.show()
```

```python
y=3*x**2+1
y
```

```python
plt.plot(x, y, 'r')
plt.xlabel('Independent Variable')
plt.ylabel('Dependent Variable')
plt.title('To plot the signal of y=3*x**2+1')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x, y, 'r--')
plt.subplot(1,2,2)
plt.plot(y, x, 'g*-')
```

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x, y, 'b')
axes.set_xlabel('Set X Label')
axes.set_ylabel('Set Y Label')
axes.set_title('Set Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.plot(x, y, 'b')
axes1.set_xlabel('Set X Label')
axes1.set_ylabel('Set Y Label')
axes1.set_title('Set Title')
axes2.plot(x, y, 'r')
axes2.set_xlabel('Set X Label')
axes2.set_ylabel('Set Y Label')
axes2.set_title('Set Title')
```

```python
fig,axes=plt.subplots()
axes.plot(x, y, 'r')
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Title');
```

```python
fig,axes=plt.subplots(nrows=1, ncols=2)
```

```python
axes
```

```python
for ax in axes:
    ax.plot(x, y, 'b')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
fig  
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
for ax in axes:
    ax.plot(x, y, 'g')
    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_title('Title')
fig
plt.tight_layout()    
```
